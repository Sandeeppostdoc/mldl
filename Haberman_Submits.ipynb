Graph for Probability density function and cumulative density function for the feature year. Observation: **PDF** : 30% of distribution are present of year 64 to 65 approximately 

**CDF**:  80% are present below 64 year approximately

```
import warnings 
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
haberman_year=haberman.loc[haberman["year"]];

counts,bin_edges=np.histogram(haberman_year['year'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf)
counts,bin_edges=np.histogram(haberman_year['year'],bins=20,density=True)
pdf=counts/(sum(counts))
plt.plot(bin_edges[1:],pdf)
plt.xlabel("Year")
plt.title('Graph for Probability density function and cumulative density function for the feature year')
plt.legend(title='CDF PDF', loc='upper left', labels=['PDF', 'CDF'])
plt.show();
```

Graph for Probability density function and cumulative density function for the feature age. Observation: **PDF** : 30% of distribution are present between age 41 to 42 approximately 

**CDF**:  80% are present below 43 age approximately

```
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
haberman_age=haberman.loc[haberman["age"]];
counts,bin_edges=np.histogram(haberman_age['age'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf)
counts,bin_edges=np.histogram(haberman_age['age'],bins=20,density=True)
pdf=counts/(sum(counts))
plt.plot(bin_edges[1:],pdf)
plt.xlabel("Age")
plt.title('Graph for Probability density function and cumulative density function for the feature Age')
plt.legend(title='CDF PDF', loc='upper left', labels=['PDF', 'CDF'])
plt.show();
```

Graph for Probability density function and cumulative density function for the feature nodes. **PDF** : There is a sudden down fall in graph for the node from 60% to 0% between values 0 to 6
 and 0% of distribution are present between nodes 15 to 28 approximately 

**CDF**:  90% are present below 25 nodes approximately 

```
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
haberman_nodes=haberman.loc[haberman["nodes"]];

warnings.filterwarnings("ignore") 
counts,bin_edges=np.histogram(haberman_nodes['nodes'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf)
counts,bin_edges=np.histogram(haberman_nodes['nodes'],bins=20,density=True)
pdf=counts/(sum(counts))
plt.plot(bin_edges[1:],pdf)
plt.xlabel("Nodes")
plt.title('Graph for Probability density function and cumulative density function for the nodes year')
plt.legend(title='CDF PDF', loc='upper left', labels=['PDF', 'CDF'])
plt.show();
```

Mean, Variance and Std-dev

```
#Mean, Variance, Std-deviation,  
print("Means:")
haberman_age=haberman.loc[haberman["age"]];
haberman_year=haberman.loc[haberman["year"]];
haberman_nodes=haberman.loc[haberman["nodes"]];

print(np.mean(haberman_age["age"]))
#Mean with an outlier.
print(np.mean(np.append(haberman_age["age"],50)));
print(np.mean(haberman_year["year"]))
print(np.mean(haberman_nodes["nodes"]))

print("\nStd-dev:");
print(np.std(haberman_age["age"]))
print(np.std(haberman_year["year"]))
print(np.std(haberman_nodes["nodes"]))
```

Median, Percentile, Quantile, IQR, MAD

```
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")

haberman_age=haberman.loc[haberman["age"]];
haberman_year=haberman.loc[haberman["year"]];
haberman_nodes=haberman.loc[haberman["nodes"]];

print(np.median(haberman_age["age"]))
#Median with an outlier
print(np.median(np.append(haberman_age["age"],50)));
print(np.median(haberman_year["year"]))
print(np.median(haberman_nodes["nodes"]))


print("\nQuantiles:")
print(np.percentile(haberman_age["age"],np.arange(0, 100, 25)))
print(np.percentile(haberman_year["year"],np.arange(0, 100, 25)))
print(np.percentile(haberman_nodes["nodes"], np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(haberman_age["age"],90))
print(np.percentile(haberman_year["year"],90))
print(np.percentile(haberman_nodes["nodes"], 90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(haberman_age["age"]))
print(robust.mad(haberman_year["year"]))
print(robust.mad(haberman_nodes["nodes"]))
```
